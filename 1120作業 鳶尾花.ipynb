```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x[:5],y[:5]
```

```python
X = x[:,:3]
Y = y
```

```python
len(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,
                                                test_size=0.5,
                                                random_state=0)
```

```python
plt.scatter(X[:,0],X[:,2],c=Y,cmap='Paired')
```

```python
from sklearn.svm import SVC
clf = SVC()
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
plt.scatter(x_test[:,0],x_test[:,2],c=y_predict-y_test)
```

```python
x_arr = np.arange(3.5,10,0.02)
y_arr = np.arange(1.0,7.0,0.02)
X,Y = np.meshgrid(x_arr,y_arr)
P = np.c_[X.ravel(),Y.ravel()]
z = clf.predict(P)
Z = z.reshape(X.shape)

plt.contourf(X,Y,Z,cmap='Paired',alpha=0.3)
plt.scatter(x[:,0],x[:,3],cmap='Paired',c=y_test)
```
