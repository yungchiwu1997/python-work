```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


def myplot(n=1):
    x=np.linspace(-5,5,500)
    y=np.sinc(n*x)
    plt.plot(x,y)
```

```python
interact(myplot,n=(1,20));
```
